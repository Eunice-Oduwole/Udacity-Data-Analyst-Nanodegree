```python
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)


api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

[Ref 1 (api = tweepy.API(auth) won't work)](https://github.com/tweepy/tweepy/issues/1102)

```python
Tweet_id = list(twitter_archive.tweet_id)

tweet_statuss = []
tweet_statuss_id= []
id_errors = []
for ids in Tweet_id:
    try:
        statuss = api.get_status(ids, 
                              tweet_mode='extended',
                              wait_on_rate_limit = True,
                              wait_on_rate_limit_notify = True)
        tweet_statuss.append(statuss)
        tweet_statuss_id.append(ids)
    except:
        id_errors.append(ids)
        print(ids)
```

```python
with open('tweet_json.txt', mode = 'w') as outfile:
    json.dump(tweet_statuss, outfile)
```

```python
# Load the Twitter API data
tweet_json = pd.read_json('tweet_json.txt')
tweet_json['tweet_id'] = tweet_statuss_id
```

#### `twitter_archive`:  Floating number not capture in  rating_numerator column 

##### Define 
Extract the floating number from the text column
[Ref 2](https://stackoverflow.com/questions/4703390/how-to-extract-a-floating-number-from-a-string), [Ref 3](https://stackoverflow.com/questions/4842956/python-how-to-remove-empty-lists-from-a-list) and then replace the affected columns with the right number

##### Code

```python
float_rating = []
rating_loc = []
rating_text = []
t=0
while t < len(twitter_archive_clean):
    if bool(re.findall('\d+\.\d+\/\d+', twitter_archive_clean.text[t])):
        float_rating.append(re.findall('\d+\.\d+', twitter_archive_clean.text[t]))
        rating_loc.append(t)
        rating_text.append(twitter_archive_clean.text[t])        
    t = t + 1
rating_text
```

```python
float_rating
```

```python
i = 0
while i < len(rating_loc):
    twitter_archive_clean.loc[rating_loc[i],'rating_numerator'] = float_rating[i]
    i= i + 1
```

##### Test

```python
twitter_archive_clean.loc[rating_loc[0]]
```

#### `twitter_archive`:  There are mistakes in the name column such as 'a', 'getting', 'this', 'all' etc.

##### Define
Locate the names using the lowercase [Ref 4](https://stackoverflow.com/questions/12934997/how-to-detect-lowercase-letters-in-python) and replace them with nan (also replace the existing 'None' in the coulmn with nan such that we are consitent

```python
names = []
for n in twitter_archive_clean.name:
    if n[0].islower() and n not in names:
        names.append(n)
names
```

```python
twitter_archive_clean.name.replace(names,np.nan,inplace = True)
twitter_archive_clean.name.replace('None',np.nan,inplace = True)
```

```python
twitter_archive_clean.name
```

```python
twitter_archive_clean.groupby(['dog_stages'])['favorite_count'].count().plot(kind='bar',figsize =(8,8));
plt.title('Bar chat of dog stages showing favorite count', fontsize = 15, fontweight = 'bold');
plt.xlabel('Dog stages');
plt.ylabel('Favorite count');
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.savefig('Favorite_count.png');
```

```python
twitter_archive_clean.groupby(['dog_stages'])['rating_numerator'].mean().plot(kind='bar',figsize =(10,8));
plt.title('Bar chat of dog stages showing average rating');
plt.xlabel('Dog stages');
plt.ylabel('Average rating');
plt.savefig('Average_rating.png');
```
