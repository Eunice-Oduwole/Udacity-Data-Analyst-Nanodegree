## Investigation Overview

 In this project, the aim is to look at the characteristics of bike hiring that could be used to predict trip duration. The main focus was on days of the week, user type (subscriber or customer i.e casual user) and hours of the day.

## Dataset Overview

 The dataset consist of trip duration and attributes of about 192082 entries. The attribute includes four feature, with other attributes such as bike_share_for_all_trip, start_station_name, start station id. I added two more features such that we have entries for time in terms of day and time (extracted from the start time column). 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from datetime import datetime
import calendar

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
tripdata = pd.read_csv('201901-fordgobike-tripdata.csv')
```

```python
tripdata.user_type = tripdata.user_type.astype('category')
tripdata.bike_share_for_all_trip = tripdata.bike_share_for_all_trip.astype('category')
tripdata.start_time = pd.to_datetime(tripdata.start_time)
tripdata.end_time = pd.to_datetime(tripdata.end_time)
tripdata.bike_id = tripdata.bike_id.astype(str)
```

```python
tripdata_copy = tripdata.copy()
```

```python
days = []
hours = []
i = 0

while i < len(tripdata):
    result = calendar.day_name[tripdata_copy.start_time[i].weekday()]
    result2 = tripdata_copy.start_time[i].hour
    days.append(result)
    hours.append(result2)
    i = i +1
tripdata_copy['day_of_the_week'] = days
tripdata_copy['hours_of_the_day'] = hours
```

```python
ordinal_var_dict = {'day_of_the_week': ['Sunday','Monday','Tuesday','Wednesday','Thursday', 'Friday', 'Saturday']}
for var in ordinal_var_dict:
    ordered_var = pd.api.types.CategoricalDtype(ordered = True,
                                                categories = ordinal_var_dict[var])
    tripdata_copy[var] = tripdata_copy[var].astype(ordered_var)
```

## Distribution of trip durations

 The trip duration had a large range of values, from about 0 at the lowest to about  7800 at the highest. Thus, a plot was made using a logarithmic scale and  the distribution appeared to be skewed to the right as shown below. 

```python
log_binsize = 0.05
bins_log = 10 ** np.arange(2.4, np.log10(tripdata_copy.duration_sec.max()) + log_binsize, log_binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = tripdata_copy, x = 'duration_sec', bins = bins_log)
plt.title('Distribution of trip durations')
plt.xlabel('Trip Duration')
plt.ylabel('Count of trips')
plt.xscale('log')
plt.xticks([500, 1e3, 2e3, 5e3, 1e4], [500, '1k', '2k', '5k', '10k'])
plt.axis([0, 10000, 0, 15000])
plt.show()
```

## Distribution of hours of the day

 The distribution of the hour of day is roughly bimodal, with the first peak roughly between 8:00 and 10:00. There is a sharp drop in the trip frequency after 10am, this could probably be because people are now settled for the day's activities such as office work, school etc. Again, the trip starts to increase at about 14:00 and reaches it's second peak at about 16:00 and gradually decreases. Again, this could be attributed to fact that people are closed for the day and take this bike trip to there various destinations. 

```python
bins = np.arange(0, tripdata_copy.hours_of_the_day.max()+2, 2)
plt.figure(figsize=[8, 5])
plt.hist(data = tripdata_copy, x = 'hours_of_the_day', bins = bins);
plt.title('Distribution of hours of the day')
plt.xlabel('Hours of the day')
plt.ylabel('Count of trips')
plt.show()
```

## Distribution of trips for days of the week

This chart reveals that more trips were taken during the working days (Monday - Friday), with Thursday having a higher trip count. 

```python
default_color = sb.color_palette()[0]
sb.countplot(data = tripdata_copy, x = 'day_of_the_week', color = default_color);
plt.title('Distribution of trips for days of the week')
plt.xlabel('Day of the week')
plt.ylabel('Count of trips')
plt.xticks(rotation = 90);
```

## Count of trips for users
The chart reveals that subscribers take more trips than the causal uers (customer).

```python
default_color = sb.color_palette()[0];
sb.countplot(data = tripdata_copy, x = 'user_type', color = default_color);
plt.title('Count of trips for user type');
plt.xlabel('User type');
plt.ylabel('Count of trips');
```

## User type versus Trip duration 
Plotting user type against the trip duartion, it is quite interesting to see that casual users(cutomer) took a longer trip duration than the subsriber who had higher trip count from the previous chart. We can probably say casual user mostly took the bike trip when they have to go a longer distance, while subcribers take the bike trip on a dialy basis. 

```python
plt.figure(figsize = [10, 5])
base_color = sb.color_palette()[0]
sb.boxplot(data = tripdata_copy, x = 'user_type', y = 'duration_sec', color = base_color)
plt.ylim([-10, 3000])
plt.title('User type and trip duration')
plt.xlabel('User type')
plt.ylabel('Trip duration')
plt.show()
```

## Trip Duration versus day of the week

This plot reveal that the longer trips occurs on Sundays and Saturdays. Probably because it's a free day for most people.  

```python
plt.figure(figsize = [10, 5])
base_color = sb.color_palette()[0]
sb.boxplot(data = tripdata_copy, x = 'day_of_the_week', y = 'duration_sec', color = base_color)
plt.ylim([-10, 3000])
plt.title('Trip Duration and day of the week')
plt.xlabel('Hours of the day')
plt.ylabel('Trip duration')
plt.show()
```

## Trip duration and  hour by days of the week
Here, I want to take a look at the interactive effect of trip duration, hours of the day and days of the week. Taking a look at this plot, it further attest to our previous observations that longer trips occurs during the weekend. 

```python
plt.figure(figsize = [10, 5])
categoric_vars = [['Sunday', '*'],['Monday','D'],['Tuesday','x'],['Wednesday','+'],['Thursday', 'o'], ['Friday','s'], ['Saturday','v']]
for day,markers in categoric_vars:
    tripdata_day = tripdata_copy[tripdata_copy['day_of_the_week'] == day]
    plt.scatter((tripdata_day['hours_of_the_day']), tripdata_day['duration_sec'], marker = markers, alpha= 1)
plt.legend(['Sunday','Monday','Tuesday','Wednesday','Thursday', 'Friday', 'Saturday'])
plt.title('Day of the week, trip duration and hours of the week')
plt.xlabel('Hours of the day')
plt.ylabel('Trip duration')
plt.show()
```
