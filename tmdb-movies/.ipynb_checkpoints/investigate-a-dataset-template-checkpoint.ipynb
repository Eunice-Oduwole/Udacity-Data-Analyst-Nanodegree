### Dropping Extraneous Columns
Below we shall columns that not relevant to the questions we wish to address in this project.

```python
df.drop(columns=['overview','homepage','tagline','original_title','release_date','keywords'],inplace= True)
# let's confirm these cloumns have been dropped 
df.columns
```

## Data Cleaning
> Here, we shall check the total number of missing vlaues in each cloumn of the dataset.
### Drop Rows with Missing Values

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum().any()
```

```python
print(df.shape)
```

## Dropping Rows with Duplicated Values

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace= True)
df.duplicated().sum().any()
```

# Dealing with values outside realistic range
To address unrealistic values which was detected from the descriptive statistics, we shall replace every row that has a zero value with 'nan'. Then find out the total numbers of such entries.

```python
# replace zeros with 'nan'

df = df.replace(0, nan)
print(df.isnull().sum())
```

For the runtime we are going to assume that the values are missing at completely as such that, the missing values are distributed similarly to values that are present (MCAR). Thus, we shall fill it with the mean value


```python
df['runtime'].fillna(df['runtime'].mean(), inplace=True)
```

```python
df.dropna(inplace=True)
df.isnull().sum().any()
```

## Inspecting and fixing data types
We are going to convert the runtime, budget and revenue from to float integer

```python
df.dtypes
```

```python
df['runtime'] = df['runtime'].astype(int)
```

### What type (genre) of movies are  produced over the years?
To achieve this, we loop over the years to count the number of each unique genre and appending it to a list and then transpose it in other to add it to the dataframe. 

```python
DF = pd.read_csv('genres.csv')
```

```python
#['Comedy', 'Action', 'Adventure', 'Western', 'Science Fiction', 'Drama', 'Family', 'Crime', 'Romance', 'War', 'Thriller','Fantasy', 'History','Mystery', 'Animation', 'Horror', 'Music', 'Documentary', 'TVMovie','Foreign']



numbers=[]
year = 1960
while year < 2016:
    d_year = df['release_year'] == year
    iscomedy = df['genres'].str.contains('Foreign')
    Y = df[iscomedy & d_year]
    G = Y['genres'].count()
    numbers.append(G)
    year = year + 1


J = np.transpose(numbers)
DF['Foreign'] = J
DF.head()
```

```python
DF.to_csv('genres_year.csv',index=False)
```

### Does movies with less runtime recieve higher vote?
>Here we try to split the runtime into high and low using the median, then we find the average vote for both. 

```python
R_mean =df['runtime'].median()
R_mean
```

```python
low_runtime =df[df['runtime']< R_mean]
high_runtime =df[df['runtime']>= R_mean]
Samples = df.shape[0]
Samples == low_runtime['vote_average'].count() + high_runtime['vote_average'].count() # should be True
```

```python
mean_low_vote = low_runtime['vote_average'].mean()
mean_high_vote =high_runtime['vote_average'].mean()
print(mean_low_vote, mean_high_vote)
```

```python
plt.figure(figsize =(10,8))
locations = [1, 2]
heights = [mean_low_vote, mean_high_vote]
labels = ['Low runtime', 'High runtime']
plt.bar(locations, heights, tick_label=labels)
plt.title('Average vote by Runtime',fontsize=20)
plt.xlabel('Runtime',fontsize=20)
plt.ylabel('Average vote',fontsize=20);
```

From the diagram, it is seen that the movies with lower runtime actually recieve lesser vote compared to those with a higer runtime.

### What kind of properties are associated with movies that have higher revenues?
>To get the movies with higher revenues, we selected those whose revenue is greater than the average revenue for the dataset. Then, we obtain a descriptive statistics for it.


```python
Rev =df['revenue'].mean()
high_revenue =df[df['revenue']>= Rev]
high_revenue.describe()
```

>From this table, it shows that on average movies with higher revenues have a budget **7.449737e+07**, runtime of **115.61** and vote_average of **6.39**
